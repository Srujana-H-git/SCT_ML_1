```python
data['bedrooms'].value_counts().plot(kind='bar') 
plt.title("number of bedrooms")
plt.xlabel('bedrooms')
plt.ylabel('count')
sns.despine
```

```python
plt.scatter(data.price,data.bedrooms)
plt.title("price vs bedrooms")
```
